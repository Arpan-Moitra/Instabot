# 4. Follow/Unfollow given handle - 
## 1. Open the Instagram Handle of “So Delhi”
## 2. Start following it. Print a message if you are already following
## 3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
"""
Follow/Unfollow given handle - 
Open the Instagram Handle of “So Delhi”
Start following it. Print a message if you are already following
After following, unfollow the instagram handle. Print a message if you have already unfollowed. 
"""

# importing all the necessary libraries
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

# initiating a webdriver session
driver = webdriver.Chrome(ChromeDriverManager().install())
    
def login(username, password):

    # sending a get request to instagrams landing page 
    driver.get("https://www.instagram.com/")

    # using explicit wait till the home page loads and username and password textboxes can be located
    un = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, 'username')))
    ps = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, 'password')))

    # passing username and password and pressing enter after passing the password
    un.send_keys(username)
    ps.send_keys(password)
    ps.send_keys(Keys.ENTER)

    # using explicit wait till the next page loads and selecting not to save password after the page has loaded
    bt1  = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME, 'cmbtv')))
    bt1.click()

    # using explicit wait till the next page loads and selecting no notifications after the page has loaded
    bt2 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "HoLwm")]')))
    bt2.click()

login(username, password)

def openAccount(account_name):
    # Searching for food in the search bar. Not using explicit wait because the page is already loaded
    sb = driver.find_element_by_xpath("//input[@placeholder = 'Search']")
    sb.send_keys(account_name)

    # Waiting for the search results to load and clicking the first search result
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "qQT3")]'))).click()

openAccount("So Delhi")

# Follows account that is already opened if account isnt open, open account first and then follow
def followAccount():
    
    #using explicit wait to wait for account to load and then searching for follow/unfollow button
    btn = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//span[@class = "vBF20 _1OSdk"]//button')))
    
    # converting into html data to check if page is already being followed or is not followed
    data = BeautifulSoup(btn.get_attribute('outerHTML'), 'html.parser')
    if data.text == 'Follow':
        btn.click()
        print('Started Following So Delhi')
    else:
        print('Already Following So Delhi')
    
followAccount()
time.sleep(5)
followAccount()

time.sleep(10)

# Unfollows account that is already opened if account isnt open, open account first and then follow
def unfollowAccount():
    
    #using explicit wait to wait for account to load and then searching for follow/unfollow button
    btn = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//span[@class = "vBF20 _1OSdk"]//button')))
    
    # converting into html data to check if page is already being followed or is not followed
    data = BeautifulSoup(btn.get_attribute('outerHTML'), 'html.parser')
    if data.text == '':
        btn.click()
        
        #using explicit wait to wait for confirmation page to load
        WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//button[@class = "aOOlW -Cab_   "]'))).click()
        print('Unfollowed So Delhi')
    else:
        print('Already Unfollowed So Delhi')
    
unfollowAccount()
time.sleep(5)
unfollowAccount()

time.sleep(10)
driver.close()
```

# 5. Like/Unlike posts
## 1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
## 2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python

# importing all the necessary libraries
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

# initiating a webdriver session
driver = webdriver.Chrome(ChromeDriverManager().install())
    
def login(username, password):

    # sending a get request to instagrams landing page 
    driver.get("https://www.instagram.com/")

    # using explicit wait till the home page loads and username and password textboxes can be located
    un = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, 'username')))
    ps = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, 'password')))

    # passing username and password and pressing enter after passing the password
    un.send_keys(username)
    ps.send_keys(password)
    ps.send_keys(Keys.ENTER)

    # using explicit wait till the next page loads and selecting not to save password after the page has loaded
    bt1  = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME, 'cmbtv')))
    bt1.click()

    # using explicit wait till the next page loads and selecting no notifications after the page has loaded
    bt2 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "HoLwm")]')))
    bt2.click()

login(username, password)

def openAccount(account_name):
    # Searching for food in the search bar. Not using explicit wait because the page is already loaded
    sb = driver.find_element_by_xpath("//input[@placeholder = 'Search']")
    sb.send_keys(account_name)

    # Waiting for the search results to load and clicking the first search result
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "qQT3")]'))).click()

openAccount("dilsefoodie")

def likeUnlikePosts(s, n):
    
    #opening the first post
    post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='eLAPa']")))
    post.click()
    time.sleep(2)
    
    i = 1
    while True:
        
        #Finding the like button
        like_button = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//span[@class='fr66n']/button")))
        
        #Finding the type of the button whether like or unlike
        type_button = BeautifulSoup(like_button.get_attribute('innerHTML'),'html.parser').find('svg').get('aria-label')
        
        # Clicking the buttonif the button is same as the command provided as funtion attribute do else printing already liked/unliked and not clicking the button
        if type_button.strip().lower()==s.strip().lower():
            like_button.click()
        
        else:
            print('Already {}d'.format(s))
            
        #Finding the next post navigator and clicking it after it has loaded
        next_post = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class=' l8mY4 feth3']/button")))
        next_post.click()
        
        i+=1 # Incrementing the post-counter
        
        # If counter is one more than number of posts to be liked printing completion message and breaking from the loop
        if i == n + 1:
            print("{}d".format(s.title()), n, "recent posts")
            
            # Closing the opened post
            driver.execute_script("arguments[0].click();", driver.find_element_by_xpath("//div[@class='NOTWr']/button"))
            break
            
            
likeUnlikePosts("like", 30)

time.sleep(2)
likeUnlikePosts("unlike", 30)

time.sleep(10)
driver.close()
```

# 6. Extract list of followers
## 1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
## 2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python

# importing all the necessary libraries
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

# initiating a webdriver session
driver = webdriver.Chrome(ChromeDriverManager().install())
    
def login(username, password):

    # sending a get request to instagrams landing page 
    driver.get("https://www.instagram.com/")

    # using explicit wait till the home page loads and username and password textboxes can be located
    un = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, 'username')))
    ps = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.NAME, 'password')))

    # passing username and password and pressing enter after passing the password
    un.send_keys(username)
    ps.send_keys(password)
    ps.send_keys(Keys.ENTER)

    # using explicit wait till the next page loads and selecting not to save password after the page has loaded
    bt1  = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME, 'cmbtv')))
    bt1.click()

    # using explicit wait till the next page loads and selecting no notifications after the page has loaded
    bt2 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "HoLwm")]')))
    bt2.click()

login(username, password)

def openAccount(account_name):
    # Searching for food in the search bar. Not using explicit wait because the page is already loaded
    sb = driver.find_element_by_xpath("//input[@placeholder = 'Search']")
    sb.send_keys(account_name)

    # Waiting for the search results to load and clicking the first search result
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH, '//*[contains(@class, "-qQT3")]'))).click()

def get_user_ids(option, page):
    
    # Opening the page
    openAccount(page)
    time.sleep(5)
    
    # Clicking on either followers or following depending on option specified
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, option))).click()
    time.sleep(3)

    current_height = 0
    i = 0
    
    # Implementing infinite scrolling till 500+ elements
    while True:
        try:
            prev_height = current_height
            
            # Finding page height/scrollheight for div class isgrP using js and scrolling to that height
            current_height = driver.execute_script('return document.querySelector(".isgrP").scrollHeight;')
            driver.execute_script('document.querySelector(".isgrP").scrollTo(0,arguments[0]);',current_height)
            time.sleep(1)
            
            # Breaking if list is exhausted and no more scrolling is happening
            if prev_height==current_height:
                break
            i += 12
            
            # Breaking if 500+ elements have been obtained 
            if i == 540:
                break
        except:
            break
    
    # Storing selenium elements for each user-id obtained in an array to retrieve usernames and store it in another array
    arr = driver.find_elements_by_xpath("//a[contains(@class, 'FPmhX')]")
    res = []
    time.sleep(2)
    for i in arr[0:500]:
        i = BeautifulSoup(i.get_attribute('outerHTML'),'html.parser')
        res.append(i.a['title'])
    
    # Closing the Follower/Following list   
    driver.find_element_by_xpath("//div[@class = 'WaOAr']/button").click()
    return res


# Getting top 500 followers of foodtalkindia
foodtalkindia_follower_ids = get_user_ids("followers","foodtalkindia")
time.sleep(3)

# Getting top 500 followers of sodelhi
sodelhi_follower_ids = get_user_ids("followers","sodelhi")
time.sleep(3)

# Getting top 500 followers or max followers of self
my_followers = get_user_ids("followers","aro.ann")
time.sleep(3)

# Getting top 500 followers or max following of self
i_am_following = get_user_ids("following","aro.ann")

time.sleep(3)
# 500 followers of foodtalkindia
print("500 followers of foodtalkindia:")
print(foodtalkindia_follower_ids)

print("---------------------------------------------------")

# 500 followers of sodelhi
print("500 followers of sodelhi:")
print(sodelhi_follower_ids)

print("---------------------------------------------------")

# Print follower of foodtalkindia who are being followed by me and they aren't following back
print("Follower of foodtalkindia who are being followed by me and they aren't following back")
for user_id in i_am_following:
    if user_id in foodtalkindia_follower_ids and user_id not in my_followers:
        print(user_id)
        
time.sleep(10)
driver.close()
```
